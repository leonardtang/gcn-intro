```python
""" Initializing Graph Convolutional Network for feature representation """

import networkx as nx
from networkx import karate_club_graph, to_numpy_matrix
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

zkc = karate_club_graph()
order = sorted(list(zkc.nodes()))  # Numbered list of nodes 

# Visualizing the network
nx.draw_circular(zkc, with_labels=True)
plt.show()
fig, ax = plt.subplots(1, 1, figsize=(8, 6));
nx.draw_networkx(zkc, ax=ax)
plt.show()

A = to_numpy_matrix(zkc, nodelist=order)
I = np.eye(zkc.number_of_nodes())

A_hat = A + I  # Adding self-loops
D_hat = np.array(np.sum(A_hat, axis=0))[0]  # Degree matrix (sum across rows)
D_hat = np.matrix(np.diag(D_hat)) 
```

```python
""" Initializing random weights (two layers) """

W_1 = np.random.normal(loc=0, scale=1, size=(zkc.number_of_nodes(), 4))  # Maps N input features into 4 output features
W_2 = np.random.normal(loc=0, size=(W_1.shape[1], 2))  # Maps 4 input features into 2 output features
```

```python
""" Forward pass through the network """

def relu(X):
    return np.maximum(0,X)

def conv(A_hat, D_hat, X, W):
    return relu(D_hat**-1 * A_hat * X * W)

# Using X = I as N x F input matrix (each node is one-hot encoded vector)
H_1 = conv(A_hat, D_hat, I, W_1)
H_2 = conv(A_hat, D_hat, H_1, W_2)
output = H_2  # Output matrix N x 2 (nodes and sub-group)
```
